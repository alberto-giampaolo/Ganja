```python
grid_pts = {} #List of grid point predictions. Each element lists epochs
for n in range(50):
    lx=[]
    for x in range(15):
        target = '/scratch/snx3000/albertog/log/scan_job%d/valid_100k_%d.hd5'%(n,x)
        try:
            lx += [pd.read_hdf(target)]
            #print(lx)
        except ValueError: 
            print('%d-%d is empty'%(n,x))
        except FileNotFoundError:
            print('%d-%d not found'%(n,x))
    if len(lx)>0: grid_pts[n] = lx
```

```python
tested_vars = ['maja', 'mina', 'ptD', 'tau1', 'tau2', 'tau21', 'tau3', 'tau32', 'total']
tested_bins = [np.linspace(0,0.3,31),np.linspace(0,0.3,31),np.linspace(0,1,31),
              np.linspace(0,20,31),np.linspace(0,20,31),np.linspace(0,1,31),
              np.linspace(0,20,31),np.linspace(0,1,31),np.linspace(0,1,31)]

tested_rings = ['ring0','ring1', 'ring2', 'ring3', 'ring4']

#min_eps_grid = []
min_eps_grid_rings = {}
for n_pt, pt in grid_pts.items():
    #min_eps = []
    #for v,b in zip(tested_vars,tested_bins):
    #    ep = minimizing_epoch(v, grid_pts[n],b)
    #    print(grid_pts[n])
    #    print("ks value minimizing epoch for",v,':', ep)
    #    min_eps +=[ep]
    #min_eps_grid += [min_eps]
    min_eps_grid_rings[n_pt]= minimizing_epoch_sum(tested_rings,grid_pts[n_pt],np.linspace(0,1,31))
```

```python
output_start=7738716

def print_params(pt_num,linestop=51, linestart=2):
    with open("slurm-%d.out"%(output_start+pt_num)) as logfile:
        head = [next(logfile) for x in np.arange(linestop)]
    for item in head[linestart:]:
        print(item)

def get_params(pt_num,linestop=51, linestart=2):
    with open("slurm-%d.out"%(output_start+pt_num)) as logfile:
        head = [next(logfile) for x in np.arange(linestop)]
    return head[linestart:]
        
def min_ks():
    point, ks = None, np.inf
    for n_pt, pt in min_eps_grid_rings.items():
        if pt[1]<ks: point, ks = n_pt, pt[1]
    return point,ks
```
